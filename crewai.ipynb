```python
from crewai import Agent
from crewai import Task
from crewai import Crew
from crewai import Process
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()
# llm = ChatOpenAI(model_name="gpt-4o-mini", temperature=0)
```

```python
writer = Agent(
    role = "Email Writer",
    goal = "Write a professional banking email",
    backstory = "You are an experienced banking communications specialist",
    llm = "gpt-4o-mini",
    verbose = True
)

write_email = Task(
    description = "Write a professional welcome email to a premium wealth management client.",
    expected_output = "A well-structured, professional email that welcomes the client and outlines the benefits of the premium wealth management services.",
    agent = writer
)

crew_write = Crew(
    agents = [writer],
    tasks = [write_email],
    process = Process.sequential,
    verbose = True
)

result = await crew_write.kickoff_async()
```

```python
reviewer = Agent(
    role = "Email Reviewer",
    goal = "Improve grammar and tone.",
    backstory = "English language expert",
    llm = "gpt-4o-mini",
    verbose = True
)

review_email = Task(
    description = "Review the email for grammar, tone, and professionalism.",
    expected_output = "A revised version of the email with improved grammar and tone.",
    agent = reviewer
)

crew_review = Crew(
    agents = [reviewer],
    tasks = [write_email, review_email],
    process = Process.sequential,
    verbose = True
)

result = await crew_review.kickoff_async()
result
```
